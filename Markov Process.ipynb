```python
import sklearn as sk
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.naive_bayes import GaussianNB
```

```python
data = pd.read_excel('Markov_Data.xlsx')
```

```python
data_df = pd.DataFrame(data)
```

```python
# Begin Naives Bayes to track flow
num_cols = len(data_df.columns)
data_df
```

```python
# function to fetch probabilites for each step

def prob(step):

    prob = []

    page_1_data = data_df[step].value_counts(normalize = True, sort = False, ascending = False, dropna = True).to_dict()
    page_1_data_df = pd.Series(page_1_data)
    print(page_1_data_df)
    
    for i in range(0, data_df[step].nunique()):    
        page_1a_prob = page_1_data_df[i]
        prob.append(page_1a_prob)
    return prob
```

```python
prob('Step One')
```

```python
# function to fetch values for each step 

def value(step):
    
    value = []
    
    page_1 = data_df[step].value_counts(normalize = False, sort = False, ascending = False, dropna = True).to_dict()
    page_1_df = pd.Series(page_1)
    print(page_1_df)
    
    for i in range(0, data_df[step].nunique()):
        page_1a = page_1_df[i]
        value.append(page_1a)
    return value
```

```python
value('Step One')
```

```python
# State Conditional Probability 

def CP(Intersection_value, Next_Step_Value, Next_Step_Prob, Current_Step_Prob):
    
    State_Prob = ((Intersection_value/Next_Step_Value)*Next_Step_Prob)/(Current_Step_Prob)
    return State_Prob
```
