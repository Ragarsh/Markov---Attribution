```python
import sklearn as sk
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.naive_bayes import GaussianNB
```

```python
# Read data from Excel file

data = pd.read_excel('Markov_data.xlsx')
data_df = pd.DataFrame(data)
data_df
```

```python
#function to return probabilities in each step

def prob1(step):
    
    page_1 = data_df[step].value_counts(normalize = True, sort = False, ascending = False, dropna = True).to_dict()
    page_1_df = pd.Series(page_1)
    df = page_1_df.to_frame()
    df.index
    prob_df = df.add_suffix('Prob').reset_index() 
    return prob_df
```

```python
#function to fetch values for each step.

def value1(step):
    
    page_1 = data_df[step].value_counts(normalize = False, sort = False, ascending = False, dropna = True).to_dict()
    page_1_df = pd.Series(page_1)
    df = page_1_df.to_frame()
    df.index
    value_df = df.add_suffix('Count').reset_index() 
    return value_df
```

```python
# State Conditional Probability. 

def CP(Intersection_value, Next_Step_Value, Next_Step_Prob, Current_Step_Prob):
    
    #State_Prob = (((Intersection_value/Next_Step_Value)*Next_Step_Prob)/(Current_Step_Prob))
    State_Prob = ((Intersection_value)/(Current_Step_Prob))
    return State_Prob
```

```python
# Extract intersection values between columns

def State_Intersections(column1, column2):
    Step_1 = data_df.groupby([column1, column2]).size()
    Step_1df = pd.DataFrame(Step_1)
    Step_1df.index
    state_interactions = Step_1df.add_suffix('Count').reset_index()
    interaction_prob = []
    
    for k in range(0, len(state_interactions)):
        interaction_sum  = state_interactions['0Count'].sum()
        interaction_prob.append(state_interactions['0Count'][k]/interaction_sum)
    
    iteraction_probdf = pd.DataFrame(interaction_prob)
    merge = pd.concat([state_interactions, iteraction_probdf], axis = 1)
    return merge
```

```python
# Extract markov probabilities. 

def Markov(current_step, next_step):
    
    Markov_Prob = []
    Intersection_value = 0
    Next_Step_Value = 0
    Next_Step_Prob = 0
    Current_Step_prob = 0
    Current_Step_Value = 0
    values = 0
    
    values = State_Intersections(current_step, next_step)
    
    for i in range(0, len(values)):
        iter1 = ''
        iter2 = ''
        iter1 = values[current_step][i]
        iter2 = values[next_step][i]
        Intersection_value = values[0][i]
        #print(iter2) 
        #print(Intersection_value)
        
        for j in range(0, len(value1(next_step))):
            if value1(next_step)['index'][j] == iter2:
                Next_Step_Value = value1(next_step)['0Count'][j]
                #print(Next_Step_Value)
            
            if prob1(next_step)['index'][j] == iter2:
                Next_Step_Prob = prob1(next_step)['0Prob'][j]
                #print(Next_Step_Prob)
        
        for k in range(0, len(prob1(current_step))):
            if prob1(current_step)['index'][k] == iter1:
                Current_Step_Prob = prob1(current_step)['0Prob'][k]
                #print(Current_Step_Prob)

        
        Markov_Prob.append( CP(Intersection_value, Next_Step_Value, Next_Step_Prob, Current_Step_Prob) )
    
    Markov_df = pd.DataFrame(Markov_Prob, columns = ['Markov Probability'])
    return Markov_df
```

```python
#Cleaning data to align in tables.

def create_tables(current_step, next_step):
    first_step_prob = []
    Source_prob = []
    
    if next_step != 0:
        intersection_table = State_Intersections(current_step, next_step) 
        markov_probabilities = Markov(current_step, next_step)
        merge = pd.concat([intersection_table, markov_probabilities], axis = 1, join_axes = [intersection_table.index])
        return merge
    
    if next_step == 0:
        for i in range(0, len(prob1(current_step))):
            for j in range(0, len(data_df[current_step])):
                if (prob1(current_step)['index'][i] ==  data_df[current_step][j]):
                    first_step_prob.append((data_df[current_step][j],j, prob1(current_step)['0Prob'][i]))
    
    Source_prob.append(first_step_prob)
    Source_prob_df = pd.DataFrame(Source_prob[0], columns = ["Source", "Index", "Markov Probability"])
    Source_prob_df = Source_prob_df.groupby(["Source", "Index", "Markov Probability"]).size()
    Source_prob_df = pd.DataFrame(Source_prob_df)
    Source_prob_df.index
    Source_prob_df1 = Source_prob_df.add_suffix('Count').reset_index()
    return Source_prob_df1          
```

```python
# The function returns an array of markov probabilities, first step of the process and index.

def data(current_step, next_step):
    key = []
    Source_key = []
    table = create_tables(current_step, next_step)
    
    if next_step != 0:
        for i in range(0, len(data_df)):
            for j in range(0, len(table)):
                if((data_df[current_step][i] == table[current_step][j]) and (data_df[next_step][i] == table[next_step][j])):
                    key.append((table['Markov Probability'][j],table[next_step][j], i))
        return key
    
    if next_step == 0:
        for i in range(0, len(table)):
            if((data_df[current_step][i] == table[current_step][i])):
                Source_key.append((table['Markov Probability'][i],table[current_step][i], i))
        return Source_key
```

```python
# The function returns attribution values for all the steps in the data set.

def attribution_output(iter_1, iter_2):
    product = 0
    output = []
    if iter_1 == 'data_0':
        for i in range(0, len(iter_1)):
            for j in range(0, len(iter_2)):
                if(iter_1[i][2] == iter_2[j][2]):
                    product = iter_1[i][0]*iter_2[j][0]
                    output.append((product, iter_1[i][1], iter_2[j][2]))
    else:
        for i in range(0, len(iter_1)):
            for j in range(0, len(iter_2)):
                if(iter_1[i][2] == iter_2[j][2]):
                    product = iter_1[i][0]*iter_2[j][0]
                    output.append((product, iter_1[i][1], iter_2[j][2]))
    
    return output
```

```python
#The function aligns the final output in a dataframe. 

def final_output():

    #Step One
    data_0 = data('Source', 0)
    data_1 = data('Source', 'Step One')
    
    #Step Two
    data_2 = data('Step One', 'Step Two')
    
    #Step Three
    data_3 = data('Step Two', 'Step Three')
    
    #Step Four
    data_4 = data('Step Three', 'Step Four')
    
    #Sep Five
    data_5 = data('Step Four', 'Step Five')
    
    src = pd.DataFrame(data_0, columns = ['Source Probability', 'Source', 'index' ])
    
    att = attribution_output(data_0, data_1)
    att_df = pd.DataFrame(att, columns = ['Step One Probability', 'Step One', 'index'])
   

    att1 = attribution_output(data_2, att)
    att1_df = pd.DataFrame(att1, columns = ['Step Two Probability', 'Step Two', 'index'])
    
    att2 = attribution_output(data_3, att1)
    att2_df = pd.DataFrame(att2, columns = ['Step Three Probability', 'Step Three', 'index'])
    
    att3 = attribution_output(data_4, att2)
    att3_df = pd.DataFrame(att3, columns = ['Step Four Probability', 'Step Four', 'index'])

    att4 = attribution_output(data_5, att3)
    att4_df = pd.DataFrame(att4, columns = ['Step Five Probability', 'Step Five', 'index'])
     
    merge = pd.concat([src, att_df, att1_df, att2_df, att3_df, att4_df], axis = 1)
    return merge
```

```python
# The most successful marketing mediums
def success():
    cols = list(data_df.columns.values)
    source = [] 
    for i in range(0,len(cols)):
        for j in range(0,len(data_df[cols[i]])):
            if data_df[cols[i]][j] == 'Thank you':
                source.append(data_df['Source'][j])


    source_df = pd.DataFrame(source, columns = ['Source'])
    return pd.DataFrame(source_df['Source'].value_counts(normalize = True, sort = False, ascending = False, dropna = True))
```

```python
#The most unsuccessful marketing mediums
def unsuccessful():
    cols = list(data_df.columns.values)
    source = [] 
    for i in range(0,len(cols)):
        for j in range(0,len(data_df[cols[i]])):
            if data_df[cols[i]][j] == 'Exit':
                source.append(data_df['Source'][j])


    source_df = pd.DataFrame(source, columns = ['Source'])
    return pd.DataFrame(source_df['Source'].value_counts(normalize = True, sort = False, ascending = False, dropna = True))
```
